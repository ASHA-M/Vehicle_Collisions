```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (15,12)
pd.set_option('display.max_columns', None)
```

# Importing data

```python
data = pd.read_csv("Motor_Vehicle_Collisions.csv")
data.head()
```

# Cleaning Column names

```python
data.columns = ["crash_date", "crash_time", "borough", "zip_code", "latitude", "longitude", "location", "on_street_name", "cross_street_name", 
                "off_street_name", "number_of_persons_injured", "number_of_persons_killed", "number_of_pedestrians_injured", "number_of_pedestrians_killed",
                "number_of_cyclists_injured","number_of_cyclists_killed", "number_of_motorists_injured", "number_of_motorists_killed", 
                "contributing_factor_vehicle_1", "contributing_factor_vehicle_2", "contributing_factor_vehicle_3", "contributing_factor_vehicle_4", 
                "contributing_factor_vehicle_5", "collison_id", "vehicle_type_code_1", "vehicle_type_code_2", "vehicle_type_code_3", "vehicle_type_code_4", 
                "vehicle_type_code_5"]
data.head()
```

```python
data.info() 
```

# Changing data types

```python
#Changing to datetime type
data["crash_date"] = pd.to_datetime(data["crash_date"])
data["crash_time"] = pd.to_datetime(data["crash_time"])
data.info()
```

# Checking NAN's

```python
nans_sum = data.isna().sum() # sum of nans
nans_percent = nans_sum/len(data)*100 #percentage of nans
pd.DataFrame(data=nans_percent, columns=["percentage"])
```

```python
data["borough"].value_counts(dropna=False)
```

```python
for col in data.columns:
    col_value_counts = data[col].value_counts(dropna=False)
    print(f"value counts for column '{col}':\n{col_value_counts}\n")
```

# Dropping Column's

```python
data = data.drop("on_street_name", axis=1)
data = data.drop("cross_street_name", axis=1)
data = data.drop("off_street_name", axis=1)
data = data.drop("contributing_factor_vehicle_3", axis=1)
data = data.drop("contributing_factor_vehicle_4", axis=1)
data = data.drop("contributing_factor_vehicle_5", axis=1)
data = data.drop("vehicle_type_code_3", axis=1)
data = data.drop("vehicle_type_code_4", axis=1)
data = data.drop("vehicle_type_code_5", axis=1)
```

# Dropping NAN rows

```python
data = data.dropna(subset=["borough"])
data = data.dropna(subset=["zip_code"])
data = data.dropna(subset=["latitude"])
data = data.dropna(subset=["number_of_persons_injured"])
data = data.dropna(subset=["number_of_persons_killed"])
data = data.dropna(subset=["contributing_factor_vehicle_1"])
data = data.dropna(subset=["contributing_factor_vehicle_2"])
data = data.dropna(subset=["vehicle_type_code_1"])
data = data.dropna(subset=["vehicle_type_code_2"])
```

```python
nans_sum = data.isna().sum() # sum of nans
nans_percent = nans_sum/len(data)*100 #percentage of nans
pd.DataFrame(data=nans_percent, columns=["percentage"])
```

```python
for col in data.columns:
    col_value_counts = data[col].value_counts(dropna=False)
    print(f"value counts for column '{col}':\n{col_value_counts}\n")
```
